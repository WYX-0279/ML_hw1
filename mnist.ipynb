```python
# download and load mnist data from https://www.openml.org/d/554
# for this tutorial, the data have been downloaded already in './scikit_learn_data'
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
data = np.array(pd.read_csv("../input/mnist-784/mnist_784.csv"))
X=data[...,0:-1]#对应的feature
Y=data[...,-1]
X=X/255 #做归一化处理
img1 = X[0].reshape(28, 28)
#plt.imshow(img1, cmap='gray')
#plt.show()
X_train, X_test, Y_train, Y_test = train_test_split(X[::10], Y[::10], test_size=1000)#划分训练集和测试集
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logic=LogisticRegression()
logic.fit(X_train,Y_train)
train_accuracy=logic.score(X_train,Y_train)
test_accuracy=logic.score(X_test,Y_test)
print('Training accuracy: %0.2f%%' % (train_accuracy*100))
print('Testing accuracy: %0.2f%%' % (test_accuracy*100))
```

#X,Y的形式：X是向量的集合，y是对应的label

```python
from sklearn.naive_bayes import BernoulliNB

bays=BernoulliNB()
bays.fit(X_train,Y_train)
train_accuracy=bays.score(X_train,Y_train)
test_accuracy=bays.score(X_test,Y_test)
print('Training accuracy: %0.2f%%' % (train_accuracy*100))
print('Testing accuracy: %0.2f%%' % (test_accuracy*100))
```

```python
from sklearn.svm import LinearSVC

svm=LinearSVC()
svm.fit(X_train,Y_train)
train_accuracy=svm.score(X_train,Y_train)
test_accuracy=svm.score(X_test,Y_test)
print('Training accuracy: %0.2f%%' % (train_accuracy*100))
print('Testing accuracy: %0.2f%%' % (test_accuracy*100))
```

```python
from sklearn.svm import SVC

svm=SVC()
svm.fit(X_train,Y_train)
train_accuracy=svm.score(X_train,Y_train)
test_accuracy=svm.score(X_test,Y_test)
print('Training accuracy: %0.2f%%' % (train_accuracy*100))
print('Testing accuracy: %0.2f%%' % (test_accuracy*100))
```
